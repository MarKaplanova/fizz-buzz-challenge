# The Fizz-Buzz 

## Implement the following program:
Go through the numbers from 1 to 100:
+ If the number is divisible by 3 and 5, only write “FizzBuzz”
+ If the number is divisible by 3, write “Fizz”
+ If the number is divisible by 5, write “Buzz”

```python
for num in range(1, 101): 
    if (num % 3 == 0) and (num % 5 == 0): 
        print(f"{num} FizzBuzz")
    elif (num % 3 == 0): 
        print(f"{num} Fizz")
    elif (num % 5 == 0): 
        print(f"{num} Buzz")
    else: 
        print(num)
```

```python
# Basic fizz-buzz challenge
def fizzbuzz(end): 
    end_aux = end + 1
    for num in range(1, end_aux): 
        if (num % 3 == 0) and (num % 5 == 0): 
            print(f"{num} FizzBuzz")
        elif (num % 3 == 0): 
            print(f"{num} Fizz")
        elif (num % 5 == 0): 
            print(f"{num} Buzz")
        else: 
            print(num)
```

```python
fizzbuzz(20)
```

```python
# More advanced version
# make flexible start, end, and the two factors

def fizzbuzz_crazy(start, end, fizz, buzz): 
    end_aux = end + 1
    for num in range(start, end_aux): 
        if (num % fizz == 0) and (num % buzz == 0): 
            print(f"{num} FizzBuzz")
        elif (num % fizz == 0): 
            print(f"{num} Fizz")
        elif (num % buzz == 0): 
            print(f"{num} Buzz")
        else: 
            print(num)
```

```python
fizzbuzz_crazy(1, 100, 2, 5) 
```
